```python
from scipy.stats import norm
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Create Gaussian distribution of data with mean of 3 and standard deviation of 0.1 using 100,000 data points.
mu = 3
sigma = 0.1
nsize = 100000

gauss_data = norm.rvs(loc=mu, scale=sigma, size=nsize)

#Plot our Gaussian distribution on a histogram.
fig, ax = plt.subplots()
ax.set_title("Gaussian Distribution, Mean = 3, STD = 0.1")
#TODO ARE WE SURE THAT RANGE SHOULD BE THIS SMALL? THIS IS VERY ZOOMED IN.
ax.set_xlim(2.95,3.05)
hist, bins, _ = plt.hist(gauss_data, bins=50);

#Create a true Gaussian probability distribution function.
gauss_pdf = norm.pdf(bins, loc=mu, scale=sigma)

#Make sure to normalize the pdf so it lines up with the raw data, then overlay it on the histogram.
plt.plot(bins, gauss_pdf / gauss_pdf.sum() * nsize, color="r")
```

```python
#Make y-axis logarithmic scale and show plot again.
ax.semilogy()
ax.grid()
fig
```
